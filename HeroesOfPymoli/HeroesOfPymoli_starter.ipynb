```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Purchasing Analysis (Total)

```python
# Determine the total number of players by counting input in the Screen Name Column
df_no_duplicated = purchase_data.drop_duplicates(subset = "SN")
total_players = df_no_duplicated["SN"].count()
print(f"The total number of players : {total_players}")
```

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# A new summary df created by extracting series, applying a funciton, and assigned to variable.
# A df formed from a dictionary containing the variable and value.
nbr_unique_items = len(purchase_data["Item ID"].unique())
average_price = round(purchase_data["Price"].mean(),2)
nbur_purchases = purchase_data["Purchase ID"].count()
total_revenue = purchase_data["Price"].sum()

# Dictonary containing variable and value
purchase_dictionary = {"Number of Unique Items" : nbr_unique_items,
                              "Average Price" : average_price,
                              "Number of Purchases" : nbur_purchases,
                              "Total Revenue" : total_revenue}

df = pd.DataFrame(purchase_dictionary, index = [0])

# Format dataframe to contain style formatting
df.style
df
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Stratefied the percent and total count of players by gender using groupby and agg.
# Percentage was calculated and added as a separate. The columns were switched around, and the highest percent was sorted
gender_demographic_df = df_no_duplicated.groupby("Gender")["Gender"].agg(["count"])
gender_demographic_df.columns = ["Total Count"]
gender_demographic_df["Percentage of Players"] = round(gender_demographic_df["Total Count"] / (df_no_duplicated["SN"]).count() * 100,2) 
gender_demographic_df = gender_demographic_df[["Total Count", "Percentage of Players"]]
gender_demographic_df["Percentage of Players"] = gender_demographic_df["Percentage of Players"].astype(str) + '%'
gender_demographic_df.sort_values(by = "Percentage of Players", ascending = False)
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender
* For normalized purchasing, divide total purchase value by purchase count, by gender
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
# List of column names for gender purchasing df
columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Normalized Totals"]

# Extract gender purchase ID, and price, obtain summary information, and assign to new df
purchase_df = purchase_data[["Gender", "Purchase ID", "Price"]]

# Use the aggregator functions groupby, and agg to calculate the summary values, and assign to new
# Values were then sorted by total purchase value in descending order.
purchase_by_gender_df = purchase_df.groupby("Gender")["Price"].agg(["count", "mean", "sum"])
purchase_by_gender_df.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
purchase_by_gender_df["Average Total Purchase per Person"] = purchase_by_gender_df["Total Purchase Value"] / purchase_by_gender_df["Purchase Count"]
pd.options.display.float_format = '${:,.2f}'.format
purchase_by_gender_df.sort_values("Total Purchase Value", ascending = False)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_series = purchase_data["Age"]

# Use cut method to assign age bins to each row of df
age_df = pd.DataFrame(pd.cut(age_series, bins = age_bins, right = True, labels = group_names))

# Added a new column with the dataframe that contains age binning information to the original df for groupby use
purchase_data["Age Bins"] = age_df

# Aggregate by and and use the agg method to summarise by count, and assign to a new df_age
df_age = purchase_data.groupby("Age Bins")["Age"].agg(["count"])
df_age.columns = ["Total Count"]

# Calculate the percent of players in each age bin, and adds to the df_age
df_age["Percentage of Players"] = df_age["Total Count"]/len(df_age["Total Count"])
df_age = df_age[["Percentage of Players","Total Count"]]

# Organize age demographic dataframe by the total number of players per age bins in descending order
# Formatted columns so percent is to 1 decimal place and also has the % character.
df_age = df_age.round({"Percentage of Players" : 1}) 
df_age["Percentage of Players"] = df_age["Percentage of Players"].astype(str) + '%'
df_age.sort_values("Total Count", ascending = False)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create a new df that summarizes purchasing information by age bin, using groupby and agg method.
dff = purchase_data.groupby("Age Bins")["Price"].agg(["count","mean","sum"])
dff.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
dff["Avg Total Purcase per Person"] = dff["Total Purchase Value"]/dff["Purchase Count"]
dff = dff.round({"Average Purchase Price" : 2, "Total Purchase Value" : 2, "Avg Total Purcase per Person" : 2})
dff = dff.sort_values(by = "Total Purchase Value", ascending = False)
dff
```

```python
#  Determine top paying players by using groupby and agg funding, and using sort_values method on count.
df_sn = purchase_data.groupby("SN")["Price"].agg(["count","mean","sum"])
df_sn = df_sn.sort_values(by = "count", ascending = False)
df_sn.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
df_sn = df_sn.round({"Average Purchase Price" : 2, "Total Purchase Value" : 2})
df_sn.head(5)
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Determine which items have the highest sales. using groupby and agg, and sorting by purchase count (descending)
df_items = purchase_data[["Item ID", "Item Name", "Price"]]
df_items = df_items.groupby(["Item ID", "Item Name"]).agg(["count", "max", "sum"])
df_items.columns = ["Purchase Count", "Item Price", "Total Purchase Value"]
df_items = df_items.sort_values(by = "Purchase Count", ascending = False)
df_items.head(5)

```

* Sort the above table by total purchase value in descending order
