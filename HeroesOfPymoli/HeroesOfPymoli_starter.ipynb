```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Booleans to select all rows in the Gender column that have either the string male, female, or other.
male_boolean = purchase_data["Gender"] == "Male"
female_boolean = purchase_data["Gender"] == "Female"
other_boolean = purchase_data["Gender"] == "Other / Non-Disclosed"


# Determine the total number of players that are either male or female.
total_males = len(purchase_data.loc[male_boolean])
total_females = len(purchase_data.loc[female_boolean])
total_other = len(purchase_data.loc[other_boolean])

# Determine percentage
total_players = len(purchase_data["Gender"])

# Calculate percentages
percent_male = round(total_males / total_players * 100, 2)
percent_female = round(total_females / total_players * 100,2)
percent_other = round(total_other / total_players * 100,2)

#  Create a series for the male, female, and other rows of the dataframe
male_info = [percent_male, total_males]
female_info = [percent_female, total_females]
other_info = [percent_other, total_other]

# create the variables that will be used in the dataframe function
table_data = [male_info, female_info, other_info]
index_labels = ["Male", "Female", "Other / Non-Disclosed"]

demograph_df = pd.DataFrame(table_data, columns = ["Percentage of Players", "Total Count"], index = index_labels)

demograph_df.style
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender
* For normalized purchasing, divide total purchase value by purchase count, by gender
* Create a summary data frame to hold the results
* Optional: give the displayed data cleaner formatting
* Display the summary data frame

```python
# List of column names for gender purchasing df
columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value", "Normalized Totals"]

# Extract gender purchase ID, and price, obtain summary information, and assign to new df
purchase_df = purchase_data[["Gender", "Purchase ID", "Price"]]


# Use the aggregator functions groupby, ad agg to calculate the summary values, and assign to new 
purchase_by_gender_df = purchase_df.groupby("Gender")["Price"].agg(["count", "mean", "sum"])
purchase_by_gender_df.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
purchase_by_gender_df["Normalized Totals"] = purchase_by_gender_df["Total Purchase Value"] / purchase_by_gender_df["Purchase Count"]
purchase_by_gender_df
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]
age_series = purchase_data["Age"]

# Use cut method to assign age bins to each row of df
age_df = pd.DataFrame(pd.cut(age_series, bins = age_bins, right = True, labels = group_names))

# Added a new column with the dataframe that contains age binning information to the original df for groupby use
purchase_data["Age Bins"] = age_df

# Aggregate by and and use the agg method to summarise by count, and assign to a new df_age
df_age = purchase_data.groupby("Age Bins")["Age"].agg(["count"])
df_age.columns = ["Total Count"]

# Calculate the percent of players in each age bin, and adds to the df_age
df_age["Percentage of Players"] = df_age["Total Count"]/len(df_age["Total Count"])
df_age = df_age[["Percentage of Players","Total Count"]]

# cound percengate column
df_age = df_age.round({"Percentage of Players" : 2}) 
df_age



```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, etc. in the table below


* Calculate Normalized Purchasing


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
# Create a new df that summarizes purchasing information by age bin, using groupby and agg method.
purchase_data.groupby("Age Bins")["Price"].agg(["count","mean","sum"])
dff = purchase_data.groupby("Age Bins")["Price"].agg(["count","mean","sum"])
dff.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
dff["Normalized Totals"] = dff["Total Purchase Value"]/dff["Purchase Count"]
dff = dff.round({"Average Purchase Price" : 2, "Total Purchase Value" : 2, "Normalized Totals" : 2})
dff = dff.sort_values(by = "Total Purchase Value", ascending = False)
dff
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#  Determine top paying players by using groupby and agg funding, and using sort_values method on count.
df_sn = purchase_data.groupby("SN")["Price"].agg(["count","mean","sum"])
df_sn = df_sn.sort_values(by = "count", ascending = False)
df_sn.columns = ["Purchase Count", "Average Purchase Price", "Total Purchase Value"]
df_sn = df_sn.round({"Average Purchase Price" : 2, "Total Purchase Value" : 2})
df_sn
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Determine which items have the highest sales. using groupby and agg, and sorting by purchase count (descending)
df_items = purchase_data[["Item ID", "Item Name", "Price"]]
df_items = df_items.groupby(["Item ID", "Item Name"]).agg(["count", "max", "sum"])
df_items.columns = ["Purchase Count", "Item Price", "Total Purchase Value"]
df_items = df_items.sort_values(by = "Purchase Count", ascending = False)
df_items

```

* Sort the above table by total purchase value in descending order

```python
# Resort the df by total purchase value to determine if other items had higher total revenue over the items with highest sales.
df_items = df_items.sort_values(by = "Total Purchase Value", ascending = False)
pd.options.display.float_format = '${:,.2f}'.format
df_items
```
